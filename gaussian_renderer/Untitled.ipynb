```python
import torch
from torch_cluster import knn

# 定义两个点集 x 和 y
x = torch.Tensor([[-1, -1], [-1, 1], [1, -1], [1, 1]]) # 源点集
y = torch.Tensor([[-1, 0], [1, 0]]) # 查询点集

# 定义批次信息（可选）
batch_x = torch.tensor([0, 0, 0, 0]) # x 的批次
batch_y = torch.tensor([0, 0]) # y 的批次

# 执行 KNN 搜索，寻找每个 y 点在 x 中的两个最近邻
k = 2
assign_index = knn(x, y, k, batch_x=batch_x, batch_y=batch_y)

print(assign_index)
```
